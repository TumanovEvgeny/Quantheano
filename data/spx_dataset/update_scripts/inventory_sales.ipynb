### Downloading SI factor

```python
import numpy as np
import pandas as pd
import datetime
from collections import defaultdict
#import blpapiwrapper as bw
import glob
import tqdm
```

```python
import glob
path = 'D:/data/spx_dataset_upd/MarketData_raw/NA/'
files = glob.glob(path + '*.csv')
tickers = list(map(lambda f: f[42:-4], files))
print(' '.join(tickers[:10]))
```

```python
' '.join(tickers)
```

```python
market_data = pd.read_csv('D:/data/spx_dataset_upd/un_spx_daily_nonadjusted_close.csv', parse_dates=['Date'], index_col='Date')
```

```python
ideal = market_data.ix[:, 0]
ideal.name = 'ideal'
```

```python
ideal
```

```python
gdf = ideal.copy()
for sym in tqdm.tqdm(tickers):
    df = pd.read_csv('D:/data/spx_dataset_upd/Features/INV_TO_SALES/' + sym + '.csv', parse_dates=['Date'], index_col='Date')
    gdf = pd.concat([gdf, df], axis=1)

```

```python
gdf_copy = gdf.copy()
```

```python
gdf = gdf.drop('ideal', axis=1)
```

```python
gdf.columns = tickers
```

```python
for col in gdf.columns:
    try:
#    print()
        gdf[col].loc[:gdf[col].dropna().index[-1]].ffill(inplace=True)
#   break
    except:
        continue
```

```python
gdf = gdf.ffill()
```

```python
gdf = gdf.loc[gdf.index <= ideal.index[-1]]
gdf = gdf.loc[gdf.index >= ideal.index[0]]
```

```python
gdf.loc[ideal.index].to_csv('D:/data/spx_dataset_upd/Features/SALINV_FACTOR.csv')
```
