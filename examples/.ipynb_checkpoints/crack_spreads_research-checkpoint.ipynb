```python
import pandas as pd

import numpy.ma as ma
import glob
import warnings
```

### Data prepare

```python
CL_POINT_VAL = 1000.
RB_POINT_VAL = 42000.
HO_POINT_VAL = 42000.
```

```python
# read the data
args = {'parse_dates' : ['Date'], 'index_col' : 'Date'}
data = {filename[22:-4] : pd.read_csv(filename, **args).rename(columns={'Close' : filename[22:-4]})\
        for filename in glob.glob('../data/crack_dataset/*.csv')}
data = pd.concat(list(data.values()), axis=1).ffill()
data.shape
```

```python
data.head(3)
```

```python
# crack = tuple(CL size, RB size, HO size)
def get_cracks(cracks_need = [(0, 1, -1)]):
    for crack_tuple in cracks_need:
        CL_size, RB_size, HO_size = crack_tuple
        data['crack:' + ''.join(str(crack_tuple))] = data['QCL#C'] * CL_POINT_VAL * CL_size +\
            data['QRB#C'] * RB_POINT_VAL * RB_size + data['QHO#C'] * HO_POINT_VAL * HO_size
```

```python
cracks_need = [(-3, 2, 1), (-1, 1, 0), (-1, 0, 1), (0, -1, 1)]
get_cracks(cracks_need)
data.head(3)
```

```python
cracks_data_columns = list(map(lambda tup: 'crack:' + ''.join(str(tup)), cracks_need))
data[cracks_data_columns].to_csv('../data/crack_dataset/cracks.csv')
```

```python
def get_returns(dataframe, type_='abs', lag=1, shift=-1):
    if lag != -shift:
        warnings.warn('lag != shift -> be carefull')
    if type_ == 'abs':
        return dataframe.diff(lag).shift(shift)        
    elif type_ == 'pct':
        return dataframe.pct_change(lag).shift(shift)  
```
